```python
import matplotlib.pyplot as plt
import librosa
from IPython.display import Audio
import IPython
import scipy
import time
import pygame
import numpy as np
filename = 'spectre-copyrighted-ncs-release.mp3'
# filename = 'smoke_bomb.wav'
y, sr = librosa.load(filename)
final = np.array([])
M = 2400
m = 2000
for i in range(0,50):
    spec = librosa.feature.melspectrogram(y=y[i*10000:i*10000+10000],sr=sr,
                                        n_fft=2048,
                                        hop_length=512,
                                        win_length=None,
                                        window='hann',
                                        center=True, 
    pad_mode='reflect', 
    power=2.0,
    n_mels=128,fmax = M,fmin = m)

    # step3 converting mel-spectrogrma back to wav file
    res = librosa.feature.inverse.mel_to_audio(spec, 
                                            sr=sr, 
                                            n_fft=2048, 
                                            hop_length=512, 
                                            win_length=None, 
                                            window='hann', 
                                            center=True, 
                                            pad_mode='reflect', 
                                            power=2.0, 
                                            n_iter=32,
                                            fmax=M,
                                            fmin=m)
    final = np.concatenate((final, res))
print(final)
Audio(data=final, rate=sr)
```

```python
plt.figure(figsize=(60, 10))
plt.plot(final[0:100000:1000])
```

```python
tempo, beat_frames = librosa.beat.beat_track(y=final, sr=sr)
```

```python
onset_env = librosa.onset.onset_strength(y=final, sr=sr)
onset = []
j = 40
for i in range (0,j):
    onset.append(onset_env[beat_frames[i]])
plt.figure(figsize=(20, 6))
plt.scatter(beat_frames[0:j],onset)
plt.plot(beat_frames[0:j],onset)
```
